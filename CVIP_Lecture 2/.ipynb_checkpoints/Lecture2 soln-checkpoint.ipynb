```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

```python
# Read in image to be masked
img_background_bgr = cv2.imread("4.png")
img_background_rgb = cv2.cvtColor(img_background_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=[30,10])
plt.imshow(img_background_rgb)
```

```python
img_mask = cv2.imread("6.png", cv2.IMREAD_GRAYSCALE)
img_mask_inv = cv2.bitwise_not(img_mask)
plt.imshow(img_mask_inv,cmap="gray")
```

```python
img_background = cv2.bitwise_and(img_background_rgb, img_background_rgb, mask=img_mask_inv)
plt.figure(figsize=[30,10])
plt.imshow(img_background)
```

```python
#fetch masking images
mask1 = cv2.imread("5.png", cv2.IMREAD_GRAYSCALE)
mask2 = cv2.imread("2.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=[20,5])
plt.subplot(121);plt.imshow(mask1,cmap='gray')
plt.subplot(122);plt.imshow(mask2,cmap='gray')
```

```python
result = cv2.bitwise_or(mask1, mask2, mask = None)
plt.imshow(result,cmap='gray')
```

```python
img_background = cv2.bitwise_and(img_background_rgb, img_background_rgb, mask=result)
plt.figure(figsize=[30,10])
plt.imshow(img_background)
```

```python
img_read = cv2.imread("8.png", cv2.IMREAD_GRAYSCALE)
retval, img_thresh = cv2.threshold(img_read, 95, 255, cv2.THRESH_BINARY)

# Show the images
plt.figure(figsize=[18,5])
plt.subplot(121); plt.imshow(img_read, cmap="gray");         plt.title("Original");
plt.subplot(122); plt.imshow(img_thresh, cmap="gray");       plt.title("Thresholded");

print(img_thresh.shape)
```
